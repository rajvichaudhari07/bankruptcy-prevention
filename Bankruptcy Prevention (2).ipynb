```python
# Input
x = bank.iloc[:,:-1]

# Target variable
y = bank.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
#Splitting the data into training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=0.80, random_state= 0)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
##Preparing models by Logistic Regression
logisticlassifier = LogisticRegression() 
```

```python
logisticlassifier
```

```python
logisticlassifier.fit(X_train, y_train)

logisticlassifier.coef_ # coefficients of features
```

```python
# after the training the model then we predict test data
y_pred = logisticlassifier.predict(X_test)
y_pred
```

```python
plt.rcParams['figure.figsize'] = (5,3)
plt.rcParams['font.size'] = 10
plt.rcParams['figure.dpi'] = 150

plt.title('Predicted Class', fontsize = 10, fontweight = 'bold', color = 'red')
sns.countplot(x = y_pred,hue= y_pred, saturation=1.2)
plt.legend(labels=['Non-Bankrupt','Bankrupt'])

plt.tight_layout()
```

```python
#testing performance of model by using confusion matrix
from sklearn.metrics import confusion_matrix

confusion_logist = confusion_matrix(y_test, y_pred)

confusion_logist
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred),annot= True, cbar= False)
```

```python
##Accuracy of a Model
# Train Accuracy

train_acc_logist = np.mean(logisticlassifier.predict(X_train)== y_train)
train_acc_logist
```

```python
# Test Accuracy

test_acc_logist = np.mean(logisticlassifier.predict(X_test)== y_test)
test_acc_logist
```

```python
from sklearn.metrics import accuracy_score

logistic_acc = accuracy_score(y_test, y_pred)
logistic_acc
```

The model has predicted all classes correctly.
